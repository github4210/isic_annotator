```python
import os
import json

decision_tree = 'static/rev3/decisiontree-backup.json'
decision_tree_out = 'static/rev3/decisiontree.json'
dtree = json.load(open(decision_tree, 'r'))
```

```python
struct_list = []
c_offset = 0


def parseTree(step, parent, color):
    global c_offset
    
    if 'value' in step.keys():
        
        step['longtitle'] = parent + step['title']
        c_offset 
        
        import copy
        
        new_c = copy.copy(color)

        for i,c in enumerate(color):
            if c == 0:
                new_c[i] = c_offset
        
        c_offset += 1
        

                
        res = [ step['value'], parent + step['title'], new_c]

        struct_list.append(res)
    
    if len(step['options']) > 0:
        for opt in step['options']:
            parseTree(opt, parent + step['title'] + ': ', color)

# http://colorschemedesigner.com/            
color_list = [[251, 0, 13],
              [236, 252, 0],
              [126, 0, 169],
              [20, 209, 0],
              [0, 140, 255]]


for n,o in enumerate(dtree[4]['options']):
    c_offset = 0
    parseTree(o, '', color_list[n])
```
